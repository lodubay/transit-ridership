```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np
```

```python
metric2016 = pd.read_excel('Metrics_0.xlsm', sheet_name='Metrics')
```

```python
column_list = list(metric2016.columns.values)
```

```python
keep_cols = ['Name','City','State','NTD ID', 'Reporter Type', 'Primary UZA\n Population', 'Agency VOMS',
             'Cost per\n Hour','Passengers per Hour','Cost per Passenger Mile',
             'Fare Revenues Earned','Total Operating Expenses','Unlinked Passenger Trips','Vehicle Revenue Hours','Passenger Miles','Vehicle Revenue Miles']

metric2016_v2 = metric2016[keep_cols]
metric2016_v2
```

```python
del_cols = []
for i in keep_cols[7:]:
    metric2016_v2[i+' Total'] = metric2016_v2.groupby(['NTD ID'])[i].transform('sum')
    del_cols.append(i)
```

```python
print(del_cols)
```

```python
metric2016_v3 = metric2016_v2.drop(columns=del_cols)
metric2016_v3
```

```python
metric2016_v3 = metric2016_v3.drop_duplicates(subset=['NTD ID'])
```

```python
metric2016_v3.drop(metric2016_v3.loc[metric2016_v3['Primary UZA\n Population']==0].index, inplace=True)
```

```python
metric2016_v3['Unlinked Passenger Trips Total'].describe()
```

```python
metric2016_v3 = metric2016_v3.rename(columns={'Primary UZA\n Population': 'Primary UZA Population', 'Cost per\n Hour Total': 'Cost per Hour Total'})
metric2016_v3.to_csv('Metrics2016.csv')
```
