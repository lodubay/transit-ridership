# Summary Time-Series Data Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pop_data = pd.read_csv('../data/UZA_population.csv', index_col='UACE Code', dtype={'UACE Code': str})
pop_data
```

```python
upt_data = pd.read_csv('../data/UPT.csv', index_col='NTD ID', dtype={'NTD ID': str, 'UACE Code': str})
upt_data
```

```python
year_cols = [str(y) for y in range(1991, 2020)]
years = list(range(1991, 2020))
agency_info = upt_data[[c for c in upt_data.columns if not c in year_cols]].copy()
agency_info
```

```python
upt_per_capita = upt_data.set_index('UACE Code')[year_cols] / pop_data[year_cols].loc[upt_data['UACE Code']]
upt_per_capita.set_index(upt_data.index, inplace=True)
upt_per_capita
```

```python
fig, ax = plt.subplots()
for i, row in upt_per_capita.iterrows():
    ax.plot(years, row, linestyle='-', color='k', alpha=0.2)
ax.set_yscale('log')
ax.set_xlabel('Year')
ax.set_ylabel('UPT Per Capita')
plt.show()
```

```python
opexp_total = pd.read_csv('../data/OpExp_Total.csv', index_col='NTD ID', dtype={'NTD ID': str, 'UACE Code': str})
opexp_total
```

```python
inflation = pd.read_csv('../data/inflation.csv', index_col=0).T
opexp_adj = opexp_total[year_cols] / np.tile(inflation[years], (opexp_total[year_cols].shape[0], 1))
opexp_adj
```

```python
fig, ax = plt.subplots()
for i, row in opexp_adj.iterrows():
    ax.plot(years, row, linestyle='-', color='k', alpha=0.2)
ax.set_yscale('log')
ax.set_xlabel('Year')
ax.set_ylabel('Total Operating Expenses (2019 Dollars)')
plt.show()
```
