# Summary Time-Series Data Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
upt_data = pd.read_csv('../data/train/UPT_percap.csv', index_col='UACE Code', dtype={'UACE Code': str})
# upt_data.replace(0, np.nan, inplace=True)
upt_data
```

```python
years = [int(y) for y in upt_data.columns]
```

```python
fig, ax = plt.subplots()
for i, row in upt_data.iterrows():
    ax.plot(years, row, linestyle='-', color='k', alpha=0.2)
ax.set_yscale('log')
ax.set_xlabel('Year')
ax.set_ylabel('UPT Per Capita')
plt.show()
```

```python
opexp_total = pd.read_csv('../data/train/OpExp_Total_infladj_percap.csv', index_col='UACE Code', dtype={'UACE Code': str})
opexp_total
```

```python
fig, ax = plt.subplots()
for i, row in opexp_total.iterrows():
    ax.plot(years, row, linestyle='-', color='k', alpha=0.2)
ax.set_yscale('log')
ax.set_xlabel('Year')
ax.set_ylabel('Total Operating Expenses (2019 Dollars)')
plt.show()
```
