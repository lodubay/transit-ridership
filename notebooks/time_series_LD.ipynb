## All data for one city

```python
def combine_city_data(city_code, dfs, names):
    """
    Generate a single DataFrame with data from all years for one city.
    """
    data_rows = [df.loc[city_code] for df in dfs]
    all_data = pd.concat(data_rows, axis=1)
    all_data.columns = names
    all_data.index.name = 'Year'
    return all_data
    
```

```python
# Import all dataframes
ynames = ['UPT_percap', 'PMT_percap', 'FARES_infladj_percap']
ydfs = [import_training_data('%s.csv' % name) for name in ynames]
portland_yvals = combine_city_data(portland_code, ydfs, ynames)
portland_yvals
```

```python
xnames = ['VRM_percap', 'VRH_percap', 'VOMS_percap', 'TotalFund_infladj_percap', 'FedFund_frac', 'StateFund_frac', 'LocalFund_frac', 'CapFund_infladj_percap', 'OpFund_infladj_percap', 'OpFund_frac', 'OpExp_Total_infladj_percap', 'OpExp_GA_frac', 'OpExp_VM_frac', 'OpExp_VO_frac']
xdfs = [import_training_data('%s.csv' % name) for name in xnames]
portland_xvals = combine_city_data(portland_code, xdfs, xnames)
portland_xvals
```

## Offset correlations

It's possible that some features, like capital spending, correlate better with ridership with some time offset (e.g., capital funding was used to build a new transitway which opened 10 years later).

```python
xdata = portland_xvals['CapFund_infladj_percap'].values
print(xdata)
offset_data = [xdata.tolist()]
offset_data += [[0] * offset + xdata[:-offset].tolist() for offset in range(1, 20)]
offset_data = np.array(offset_data).T
print(offset_data)
```

```python
y = portland_yvals['UPT_percap'].values.ravel()
y
```

```python
from sklearn.linear_model import LinearRegression

mlr = LinearRegression()
coefs = []
for i in range(offset_data.shape[1]):
    mlr.fit(offset_data[:,i:i+1], y)
    coefs.append(mlr.coef_)

coefs
```

```python
plt.plot(np.array(coefs).flatten())
plt.xlabel('Years Offset')
plt.ylabel('Coefficient')
plt.title('CapFund_infladj_percap')
plt.show()
```

```python
# Now with operational funding
xdata = portland_xvals['OpFund_infladj_percap'].values
offset_data = [xdata.tolist()]
offset_data += [[0] * offset + xdata[:-offset].tolist() for offset in range(1, 20)]
offset_data = np.array(offset_data).T
```

```python
mlr = LinearRegression()
coefs = []
for i in range(offset_data.shape[1]):
    mlr.fit(offset_data[:,i:i+1], y)
    coefs.append(mlr.coef_)
coefs = np.array(coefs).flatten()
```

```python
plt.plot(coefs)
plt.xlabel('Years Offset')
plt.ylabel('Coefficient')
plt.title('OpFund_infladj_percap')
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

def offset_regression(city, feature, ycol='UPT_percap'):
    dfs = [import_training_data('%s.csv' % name) for name in [ycol, feature]]
    data = combine_city_data(city, dfs, [ycol, feature])
    x = data[feature].values
    y = data[ycol].values.ravel()
    # Make offset X data
    Xoffset = [x.tolist()]
    Xoffset += [x[:-offset].tolist() for offset in range(1, 25)]
    yoffset = [y.tolist()]
    yoffset += [y[:-offset].tolist() for offset in range(1, 25)]
    # Regression on each individual offset
    pipe = Pipeline([
        ('Scale', StandardScaler()),
        ('SLR', LinearRegression())
    ])
    coefs = []
    for i in range(len(Xoffset)):
        pipe.fit(np.array(Xoffset[i])[:,np.newaxis], np.array(yoffset[i])[:,np.newaxis])
        coefs.append(pipe['SLR'].coef_)
    return np.array(coefs).flatten()
    
```

```python
feature = 'CapFund_infladj_percap'
coefs = offset_regression(portland_code, feature)
plt.plot(coefs)
plt.xlabel('Years Offset')
plt.ylabel('Coefficient')
plt.title(feature)
plt.show()
```

```python
feature = 'CapFund_infladj_percap'
coefs = offset_regression(columbus_code, feature)
plt.plot(coefs)
plt.xlabel('Years Offset')
plt.ylabel('Coefficient')
plt.title(feature)
plt.show()
```

```python
fig, ax = plt.subplots()
feature = 'CapFund_infladj_percap'
ycol = 'UPT_percap'
ydf = import_training_data('%s.csv' % ycol)
counts = np.zeros(25)
for city in ydf.index:
    coefs = offset_regression(city, feature, ycol=ycol)
    counts[coefs.argmax()] += 1
    # ax.plot(np.abs(coefs) / coefs.max(), c='k', alpha=0.2)
ax.plot(counts)
ax.set_xlabel('Years Offset')
ax.set_ylabel('# max coefs')
# ax.set_ylim((0, 1))
ax.set_title(feature)
plt.show()
```
